**Dataframe**

A DataFrame is a tabular data structure that organizes data into rows and columns, creating a two-dimensional layout. Within a DataFrame, it is common to encounter columns with heterogeneous data types, meaning that different columns can contain data of varying types.

1. A DataFrame is like a table or spreadsheet with rows and columns.
2. Columns in a DataFrame can have various data types (float64, int, bool).
3. Each DataFrame column is a Series.
4. DataFrame size can change; you can add or remove columns.
5. It has labeled rows and columns.
6. Think of it as a dictionary of Series with indexed rows and columns.

```python
import pandas as pd

# Creating a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 22, 28],
        'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']}
df_sample = pd.DataFrame(data)

# Display the DataFrame
df_sample
```

*Data cleaning and preprocessing are essential steps to ensure that your data is accurate and ready for analysis. Pandas offers tools for handling missing values, dealing with duplicates, and transforming data. Handling missing values involves employing various techniques, each with its unique strengths and weaknesses. In this context, we explore filling missing values through forward and backward methods*:

* Dropping Missing Values: Use the dropna() method to remove rows or columns with missing values.
* Filling with Zeros: Fill missing values with zeros.
* Filling with Test Statistics: Replace missing values with appropriate test statistics.
* Forward or Backward Filling: Utilize the pad or fill option to fill values forward and the bfill or backfill option to fill values backward.

```python
df['ADDRESSLINE2'].dropna()  # Remove rows with missing values
df['ADDRESSLINE2'].fillna('Not Available')  # Fill missing values with a specific value
```

# **Topic 6: Merging and Joining Datasets**

Often, you'll need to combine data from multiple sources to perform more comprehensive analysis. Pandas provides methods for merging and joining datasets.

```python
# Creating two DataFrames
data1 = {'ID': [1, 2, 3],
         'Score': [85, 90, 78]}
data2 = {'ID': [2, 3, 4],
         'Grade': ['A', 'B', 'C']}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
# Inner join
merged_inner = pd.merge(df1, df2, on='ID', how='inner')

# Display the merged DataFrames
print(merged_inner)
```

```python
# Left join
merged_left = pd.merge(df1, df2, on='ID', how='left')

# Display the merged DataFrames
print(merged_left)
```
